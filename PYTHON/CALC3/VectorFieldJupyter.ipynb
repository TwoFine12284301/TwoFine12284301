```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as matplotlib

#Defining points and vectors
points = np.array([[1,2],[0,0],[2,1]])
vectors = np.array([[-10,-8],[0,0],[-8,-10]])

#Create quiver plot
matplotlib.figure()
matplotlib.quiver(points[:,0],points[:,1], vectors[:,0],vectors[:,1],angles='xy', scale_units='xy',scale=1,color=['blue','red','yellow'])

#set limits of plot
matplotlib.xlim(-10,10)
matplotlib.ylim(-10,10)

#label points
for i, (x,y) in enumerate(points):
    matplotlib.text(x,y, f'f{x},{y}',fontsize=12)

#add labels and titles
matplotlib.xlabel('x')
matplotlib.xlabel('y')
matplotlib.title('Vector Field')
matplotlib.grid()
matplotlib.show()
```
